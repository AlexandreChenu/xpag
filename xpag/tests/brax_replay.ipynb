```python
import os
import numpy as np
from IPython.display import HTML
import gym
from brax import envs
from brax.physics.base import QP
from brax.io import html
```

```python
prefix = os.path.join(os.path.expanduser("~"), "results", "xpag")
dirfiles = os.listdir(prefix)
dirfiles.sort()
dirfiles
load_dir = os.path.join(prefix, dirfiles[-1])
load_dir
```

```python
env_name = str(np.loadtxt(os.path.join(load_dir, 'episode', 'env_name.txt'), dtype='str'))
qp_pos = np.load(os.path.join(load_dir, 'episode', 'qp_pos.npy'))
qp_rot = np.load(os.path.join(load_dir, 'episode', 'qp_rot.npy'))
qp_vel = np.load(os.path.join(load_dir, 'episode', 'qp_vel.npy'))
qp_ang = np.load(os.path.join(load_dir, 'episode', 'qp_ang.npy'))
```

```python
env = envs.create(env_name=env_name)
episode_length = len(qp_pos)
episode = [QP(qp_pos[i], qp_rot[i], qp_vel[i], qp_ang[i]) 
           for i in range(episode_length)]
```

```python
HTML(html.render(env.sys, episode))
```
